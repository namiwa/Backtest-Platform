```python
import pandas as pd
import numpy as np
import math
import get_cn_data as gcd
import matplotlib.pyplot as plt
```

```python
ticker_name = '600729'
getter = gcd.stock_data(ticker_name)
ticker = getter.get_ticker()
getter.get_stock_price()
```

```python
df = pd.read_csv("/Users/zhouzijian/myproject/Backtest-Platform/cn_intraday/" + ticker + '.csv')
df.head(10)
```

```python
for i in range(10):
    df['move' + str(i)] = df['close'].shift(i)
df.head(10)
```

```python
df['m5'] = df[['move0', 'move1', 'move2', 'move3', 'move4']].sum(axis = 1, skipna = False) / 5
df.head(10)
```

```python
df['m10'] = df[['move0', 'move1', 'move2', 'move3', 'move4','move5', 'move6', 'move7', 'move8', 'move9']].sum(axis = 1, skipna = False) / 10
```

```python
df.head(10)
```

```python
df['delta'] = df['m5'] - df['m10']
df.head(20)
```

```python
df['delta1'] = df['delta'].shift(1)
df['delta2'] = df['delta'].shift(2)
df.head(20)
```

```python
df['alpha'] = df[['delta', 'delta1', 'delta2']].sum(axis = 1, skipna = False) / 3
df.head(20)
```

```python
capital = 1000000
df['capital_alloc'] = df['alpha'] * capital
df['pct_change'] = df['close'].pct_change()
df['daily_pnl'] = df['capital_alloc'] * df['pct_change']
df.head(20)
```

```python
plt.plot(df['daily_pnl'].cumsum())
```

```python
sharpe = np.sqrt(252) * df['daily_pnl'].mean() / df['daily_pnl'].std()
```
